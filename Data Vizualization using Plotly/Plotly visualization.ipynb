```python
import matplotlib.pyplot as plt

# Example data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales_A = [100, 120, 90, 110, 130]  # Sales data for product A
sales_B = [80, 100, 110, 95, 120]    # Sales data for product B

# Plotting the area chart
plt.figure(figsize=(10, 6))

# Plotting the area for product A
plt.fill_between(months, sales_A, color='skyblue', alpha=0.4, label='Product A')

# Plotting the area for product B
plt.fill_between(months, sales_B, color='salmon', alpha=0.4, label='Product B')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend()

# Display the plot
plt.show()
```

```python
def displayPathtoPrincess(n, grid):
    princess = []
    bot = []
    for k, i in enumerate(grid):
        if "p" in i:
            princess = [i.index("p"), k]
        if "m" in i:
            bot = [i.index("m"), k]
    x = princess[0] - bot[0]
    y = princess[1] - bot[1]
    if x > 0:
        for i in range(x):
            print("RIGHT")
    if x < 0:
        for i in range(-1 * x):
            print("LEFT")
    if y > 0:
        for i in range(y):
            print("DOWN")
    if y < 0:
        for i in range(-1 * y):
            print("UP")

def main():
    # Example grid input
    n = 3
    grid = [
        ["-", "-", "-"],
        ["-", "m", "-"],
        ["p", "-", "-"]
    ]
    
    # Call the function to display the path to princess
    displayPathtoPrincess(n, grid)

if __name__ == "__main__":
    main()
```
