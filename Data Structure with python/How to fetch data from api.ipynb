```python
from scipy.stats import ttest_ind

# Assuming df_campaign1 and df_campaign2 are DataFrames containing data for two different marketing campaigns
clicks_campaign1 = select_relevent_columns['sale_amount']
clicks_campaign2 = select_relevent_columns['payout']

# Perform t-test
t_statistic, p_value = ttest_ind(clicks_campaign1, clicks_campaign2)

# Check if the p-value is less than the significance level (e.g., 0.05) to determine statistical significance
if p_value < 0.05:
    print("The observed difference in clicks between the two campaigns is statistically significant.")
else:
    print("There is no statistically significant difference in clicks between the two campaigns.")
```
