# Merging 

```python
import pandas as pd
```

```python
frame_1 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'price': [12,11,44,21,23]})
```

```python
frame_2 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue']})
```

```python
frame_1
```

```python
frame_2
```

```python
pd.merge(frame_1,frame_2)
```

```python
frame_3 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue'],
                        'Brand' : ['OMG', 'ABC', 'ABC', 'POD', 'POD']})
```

```python
pd.merge(frame_1, frame_3, on = 'id')
```

```python
pd.merge(frame_1, frame_3, on = "id", how='right')
```

```python
frame_4 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue'],
                       'Brand' : ['OMG', 'POD', 'ABC', 'POD', 'ABC'],
                       'Sid' :['pencil', 'ball', 'pencil', 'pen','pen']})
```

```python
pd.merge(frame_3, frame_4, on=['id', 'Brand'], how= 'outer')
```

```python
pd.merge(frame_3, frame_4, right_index= True, left_index=True)
```

```python
frame_3.columns = ['id_2', 'color_2', 'Brand_2']
```

```python
frame_4.join(frame_3)
```
