```python
import pandas as pd
```

```python
frame_1 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'price': [12,11,44,21,23]})
```

```python
frame_2 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue']})
```

```python
frame_1
```

```python
frame_2
```

```python
pd.merge(frame_1,frame_2)
```

```python
frame_3 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue'],
                        'Brand' : ['OMG', 'ABC', 'ABC', 'POD', 'POD']})
```

```python
pd.merge(frame_1, frame_3, on = 'id')
```

```python
pd.merge(frame_1, frame_3, on = "id", how='right')
```

```python
frame_4 = pd.DataFrame({'id' :['ball', 'pencil', 'pen', 'mug', 'astray'], 
                        'color': ['white','red','red','black','blue'],
                       'Brand' : ['OMG', 'POD', 'ABC', 'POD', 'ABC'],
                       'Sid' :['pencil', 'ball', 'pencil', 'pen','pen']})
```

```python
pd.merge(frame_3, frame_4, on=['id', 'Brand'], how= 'outer')
```

```python
pd.merge(frame_3, frame_4, right_index= True, left_index=True)
```

```python
frame_3.columns = ['id_2', 'color_2', 'Brand_2']
```

```python
frame_4.join(frame_3)
```

```python
import json
import pandas as pd

# Load the JSON data
file_path = r"C:\Users\mk744\OneDrive - Poornima University\Desktop\responses.json"
with open(file_path, 'r') as file:
    data = json.load(file)

# Inspect the data structure
print(type(data))
if isinstance(data, dict):
    print(data.keys())
else:
    print("Not a dictionary")
print(data)

# Depending on the structure, preprocess the data before converting to DataFrame
if isinstance(data, dict):
    # If the JSON data is a dictionary, further inspect each key-value pair
    for key, value in data.items():
        print(f"{key}: {type(value)}")
    
    # Normalize the dictionary data
    try:
        df = pd.json_normalize(data)
    except Exception as e:
        print(f"Error normalizing data: {e}")
        df = pd.DataFrame()
elif isinstance(data, list):
    # If the JSON data is a list of dictionaries
    if all(isinstance(item, dict) for item in data):
        df = pd.json_normalize(data)
    else:
        print("List contains non-dictionary items")
        df = pd.DataFrame(data)
else:
    raise ValueError("Unexpected JSON format")

# Display the DataFrame
print(df)
```
