```python
# convert scientific value into numeric values
if 'Sale Amount' in df.columns and 'Payout' in df.columns:
    df[['Sale Amount', 'Payout']] = df[['Sale Amount', 'Payout']].astype('float')
    df[['Sale Amount', 'Payout']] = df[['Sale Amount', 'Payout']].applymap(lambda x: '{:.0f}'.format(x))
    df[['Sale Amount', 'Payout']] = df[['Sale Amount', 'Payout']].apply(pd.to_numeric)
```
