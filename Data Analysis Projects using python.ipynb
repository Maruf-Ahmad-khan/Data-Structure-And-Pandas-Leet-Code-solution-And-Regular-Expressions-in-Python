```python
import pandas as pd

def Select_cols(df: pd.DataFrame) -> pd.DataFrame:
    """
    Groups the DataFrame by 'Adv. Campaign', sums the 'Payout' values, sorts them in descending order,
    and returns the top 10 entries.
    """
    return df.groupby(['Adv. Campaign'], as_index=False)[['Payout', 'Time']].sum().sort_values(by='Payout', ascending=False).head(10)

def main() -> pd.DataFrame:
    # Reading the Excel file
    df = pd.read_excel(r"C:\Users\mk744\Downloads\Power BI project\Final Report.xlsx", sheet_name='All Consolidated Data ')
    
    # Convert 'Value' column to integer format if it exists
    if 'Payout' in df.columns:
        # Ensure the 'Value' column is treated as a float first, to handle scientific notation
        df['Payout'] = df['Payout'].astype(float)
        # Convert the float to an integer string without scientific notation
        df['Payout'] = df['Payout'].apply(lambda x: '{:.0f}'.format(x))
        # Convert the formatted string back to an integer
        df['Payout'] = df['Payout'].astype(int)
    
    # Perform the grouping and summarizing operation
    top_payouts = Select_cols(df)
    
    return top_payouts

if __name__ == "__main__":
    result = main()
    print("Top 10 Payout DataFrame:")
    print(result)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def Select_cols(df: pd.DataFrame) -> pd.DataFrame:
    """
    Groups the DataFrame by 'Adv. Campaign', sums the 'Payout' values, sorts them in descending order,
    and returns the top 10 entries.
    """
    return df.groupby(['Adv. Campaign'], as_index=False)[['Payout', 'Clicks']].sum().sort_values(by='Payout', ascending=False).head(10)

def main() -> pd.DataFrame:
    # Reading the Excel file
    df = pd.read_excel(r"C:\Users\mk744\Downloads\Power BI project\Final Report.xlsx", sheet_name='All Consolidated Data ')
    
    # Convert 'Value' column to non-scientific notation if it exists
    if 'Payout' in df.columns:
        df['Payout'] = df['Payout'].apply(lambda x: '{:.0f}'.format(x) if pd.notnull(x) else x)
        df['Payout'] = df['Payout'].astype(int)
    
    # Perform the grouping and summarizing operation
    top_payouts = Select_cols(df)
    
    return top_payouts

if __name__ == "__main__":
    result = main()
    print("Top 10 Payout DataFrame:")
    print(result)
```
