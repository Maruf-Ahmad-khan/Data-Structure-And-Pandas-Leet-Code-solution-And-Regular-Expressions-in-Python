# Top 10 Highest Clicks and Payout

```python
import pandas as pd

def Pivot_table(df: pd.DataFrame) -> pd.DataFrame:
    return df.pivot_table(index='Campaign_Name', columns='Time', values=['Clicks', 'Payout'], aggfunc='sum').fillna(0)

def main() -> pd.DataFrame:
    df = pd.read_excel(r"C:\Users\mk744\Downloads\Admitad_XO Data From Jan 2024\Consolidate data Jan 2024 to March 2024.xlsx")
    data = Pivot_table(df)
    
    # Summing clicks and payout for each Campaign_Name
    data['Total_Clicks'] = data['Clicks'].sum(axis=1)
    data['Total_Payout'] = data['Payout'].sum(axis=1)
    
    # Sorting based on Total_Clicks and Total_Payout
    sorted_data_clicks = data.sort_values(by='Total_Clicks', ascending=False)
    sorted_data_payout = data.sort_values(by='Total_Payout', ascending=False)
    
    # Selecting top 10 Campaign_Name based on clicks and payout
    top_10_clicks = sorted_data_clicks.head(10)
    top_10_payout = sorted_data_payout.head(10)
    
    return top_10_clicks, top_10_payout

if __name__ == "__main__":
    top_10_clicks, top_10_payout = main()
    print("Top 10 Campaigns based on Clicks:")
    top_10_clicks
    
```

# Trends On Graph

```python
import pandas as pd
import matplotlib.pyplot as plt

def Pivot_table(df: pd.DataFrame) -> pd.DataFrame:
    return df.pivot_table(index='Campaign_Name', columns='Time', values=['Clicks', 'Payout'], aggfunc='sum').fillna(0)

def main() -> pd.DataFrame:
    df = pd.read_excel(r"C:\Users\mk744\Downloads\Admitad_XO Data From Jan 2024\Consolidate data Jan 2024 to March 2024.xlsx")
    data = Pivot_table(df)
    
    # Summing clicks and payout for each Campaign_Name
    data['Total_Clicks'] = data['Clicks'].sum(axis=1)
    data['Total_Payout'] = data['Payout'].sum(axis=1)
    
    # Sorting based on Total_Clicks and Total_Payout
    sorted_data_clicks = data.sort_values(by='Total_Clicks', ascending=False)
    sorted_data_payout = data.sort_values(by='Total_Payout', ascending=False)
    
    # Selecting top 10 Campaign_Name based on clicks and payout
    top_10_clicks = sorted_data_clicks.head(10)
    top_10_payout = sorted_data_payout.head(10)
    
    # Visualization for top 10 campaigns based on clicks
    plt.figure(figsize=(10, 6))
    plt.bar(top_10_clicks.index, top_10_clicks['Total_Clicks'], color='Darkblue')
    plt.title('Top 10 Campaigns based on Clicks', weight = 'bold')
    plt.xlabel('Campaign Name', weight = 'bold')
    plt.ylabel('Total Clicks', weight = 'bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Visualization for top 10 campaigns based on payout
    plt.figure(figsize=(10, 6))
    plt.bar(top_10_payout.index, top_10_payout['Total_Payout'], color='Darkgreen')
    plt.title('Top 10 Campaigns based on Payout', weight = 'bold')
    plt.xlabel('Campaign Name', weight = 'bold')
    plt.ylabel('Total Payout', weight = 'bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return top_10_clicks, top_10_payout

if __name__ == "__main__":
    top_10_clicks, top_10_payout = main()
```
